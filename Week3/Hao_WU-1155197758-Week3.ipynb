## Question 1

```python
from pandas_datareader import data as pdr
import pandas as pd

import yfinance as yf
yf.pdr_override() # <== that's all it takes :-)

dict = {'Tencent':'0700.HK',
        'Alibaba':'9988.HK',
        'Meituan':'3690.HK',
        'HSBC':'0005.HK',
        'CCB':'0939.HK',
        'HKEX':'0388.HK',
        }

dfs = []
for i,j in dict.items():
        s=pdr.get_data_yahoo(j, interval='1mo', start="2021-12-01", end="2023-07-30")['Adj Close']
        df = pd.DataFrame({'Date': s.index, 'Adj_Close': s.values, 'Ticker': i})
        df['Year'] = s.index.year
        df['Month'] = s.index.month
        df['Stock_Symbol'] = j
        df['Monthly_Return'] = df['Adj_Close'].pct_change()
        df.dropna(inplace=True)
        dfs.append(df)
raw_data = pd.concat(dfs, ignore_index=True)
data = raw_data.groupby('Ticker')
for key, group in data:
    print("Group:", key)
    print(group)
    print()
```

## Question 2

```python
sorted = raw_data.sort_values('Monthly_Return', ascending=False).groupby('Ticker')
for key, group in sorted:
    print("Group:", key)
    print(group)
    print()
```

```python
sorted2 = data.apply(lambda x: x.sort_values('Monthly_Return', ascending=False))
sorted2
```

```python
best_stock = sorted['Monthly_Return'].mean().idxmax()
best_return = sorted['Monthly_Return'].mean().max()
print("Best stock is: ", best_stock)
print("Average monthly return is: ", best_return)
```

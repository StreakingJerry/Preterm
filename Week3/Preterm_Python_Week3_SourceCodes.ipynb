### Why Pandas?

Pandas is the most widely used python library in data science. The name is derived from "panel data", an econometrics term used to describe datasets that include observations over multiple time periods for the same individuals. It is a high-performance tool to extract, clean, transform and analyze data. - https://en.wikipedia.org/wiki/Pandas_(software)

NumPy provides essential features for working with clean, well-organized data. However, in real world we rarely come across such data. If NumPy lacked the flexibility of lists, Pandas core data structure called `DataFrame` provides just that. Built on top of the NumPy package, Pandas provides an efficient implementation of multidimensional arrays with row and column labels and allows heterogeneous data types as well as missing data. 

Pandas also provides efficient solution to those instances where element-wise broadcasting is not applicable such as grouping, pivots, etc. For brief summary on Pandas package, type `pd?` in a jupyter cell.

## Data Structures

Pandas has three data structures -  
- `Series` - an array of 1 dimension (similar to column in excel)
- `DataFrame` - a table of 2 dimension (similar to sheet in excel)
- `Panel` - data of 3 dimension (similar to multiple sheets in excel)

Of these, we will focus on `Series` and `DataFrame` only.


### Series

- `Series` is used to model one dimensional data. 
- It is somewhat similar to Python's dictionary in that it provides a mapping from a collection of keys to a collection of values. 

```python
# create two series 
a = pd.Series([10,20,30,40])
b = pd.Series([22,33,44,55])
display(a, b) 
```

### DataFrame

- `DataFrame` is a collection of `Series`. So, naturally it models two-dimensional data. 
- It also has Python's dictionary like interface, except for the key has multiple values, one for each `Series`.

```python
# create a dataframe based on two series
ab = pd.DataFrame( {'col_a':a, 'col_b':b} )
ab 
```

```python
aseries = aapl['Open']
aseries[1]
```

### Indexing <font color='blue'> DataFrame </font>  using <font color='red'> index numbers </font>
- To index __rows__:
    - It is not possible to index __a single__ row of a `DataFrame` without a slicing operator. 
    - Naturally, this method can also be used to index __multiple__ rows. 

```python
aapl.head(10)
```

```python
aapl[:1]
```

```python
aapl[2:3]
```

```python
aapl[1:5]
```

```python
aapl[1:10:2]
```

- To index __columns__:
    - It is not possible to index columns only using index numbers. A `Series` inside a `DataFrame` acts as a two dimensional object. The first dimension is the number of rows and the second dimension is the number of columns. However, isolated from the `DataFrame`, it has only one dimension, which is the number of rows. 

```python
aseries = aapl['Open']
print(aseries.shape)
print(aapl.shape)
```

- To index __rows and columns__:
    - `pd.iloc` method must be used. 
    - Comma can be used to separate each dimension.
    - Regular python indexing convention can be followed in each dimension.  

```python
aapl.iloc[1:10:2, 1:5:2]
```

```python
aapl.iloc[1:10:2, 4]
```

### Indexing <font color='blue'> DataFrame </font> using <font color='red'> index labels </font>
- To index __rows__:
    - `pd.loc` method must be used. 
    - To extract __a single__ row, place the row label inside the indexing operator.
    - To extract __multiple__ rows, specify row labels inside the indexing operator using either a slicing operator or a list. 

```python
aapl.loc['2022-08-01']
```

```python
aapl.loc['2022-08-01':'2022-08-05']
```



__Note__ that the row labels for this dataset is a `Timestamp` object. Thus, a list of `Timestamp` object (not string) must be passed. Data type are very important to understand as they define the operations that can be performed! 

```python
aapl.loc[[dt.date(2022, 8, 1),
           dt.date(2022, 8, 2),
           dt.date(2022, 8, 5)]]
```

- To index __columns__:
    - To extract __a single__ column, place the column name inside the indexing operator.
    - To extract __multiple__ columns, place the index labels in a list inside the indexing operator.

```python
aapl['Open']
```

```python
aapl[['Open','Close','Adj Close']]
```

- To index __rows and columns__: 
     - `pd.loc` method must be used. 
     - Comma can be used to separate each dimension.
     - Any combination of the conventions described above for index labels can be used to index each dimension.

```python
aapl.loc['2022-08-01':'2022-08-03', 'Open':'Adj Close']    # rows and columns using slicing operator
```

```python
aapl.loc[[dt.date(2022, 8, 1),dt.date(2022, 8, 2)],       # rows and columns using a list
          ['Open','Adj Close']]
```

```python
aapl.loc['2022-08-01':'2022-08-03', ['Open','Adj Close']] # rows using slicing operator and columns using list
```

### Condtional indexing

- It is often useful to select rows based on certain condition(s). 

For example, to select rows with 'Adj Close' value of more than 100 only, we would do it as follows:

```python
subdf = aapl.loc[aapl['Adj Close'] >150, ['Adj Close', 'Volume']]
display(subdf.shape, 
        subdf.head())
```

Now, to select rows with 'Adj Close' value of more than 100 and less than and equal to 200 only, we would do it as follows:

```python
subdf = aapl.loc[(aapl['Adj Close'] >150) & (aapl['Adj Close'] <=200), ['Adj Close', 'Volume']]
print(subdf.shape)
subdf.head()
```

__NOTE:__ Methods of a `Series` and `DataFrame` are available to their slices and copies as well. 

```python
aapl.loc['2022-08-01' : '2022-08-15', 'Close':'Volume'].max()
```

```python
aapl.loc['2022-08-01' : '2022-08-15', 'Close':'Volume'].max()
```

__NOTE:__ Slice of `Series` and `DataFrame` generate their views only. Therefore, any changes to the slice will change the original `Series` and `DataFrame`. If you wish not to change the original `Series` and `DataFrame`, create a __copy__ instead of a slice. Refer to NumPy module for more information on difference between view and copy.  

```python
aslice = aapl.loc['2022-08-1':'2022-08-5', 'Close':'Volume'].copy()
aslice
```

### Concat
 - `pd.concat` combines Pandas `Series` and `DataFrame` along the specified axis.

```python
from pandas_datareader import data as pdr

import yfinance as yf
yf.pdr_override() # <== that's all it takes :-)

# download dataframe
spy = pdr.get_data_yahoo("SPY", start="2022-07-30", end="2023-07-30")
spy
```

```python
pd.concat([spy, aapl])            # default axis=0
```

- Notice that the index now has duplicates. `pd.concat` also provides a few way to deal with this.

```python
try:
    pd.concat([spy ,aapl], verify_integrity=True)
except Exception as ex:
    print("!!!Error: ", ex)
```

```python
pd.concat([spy ,aapl], keys=['SPY', 'AAPL'])
```

###  Merge

 - `pd.merge` method is an entry point for database-style join.
 - The default join is inner join unless specified otherwise.

```python
display(aapl.shape, 
        spy.shape , 
        aapl.merge( spy, left_on='Date', right_on='Date').shape)
```

```python
outerjoin_df = spy.merge(aapl, how='outer')
display(outerjoin_df.shape,
        outerjoin_df)
```

- `pd.merge` in general disregards index unless specified.

```python
spy.merge(aapl, left_index=True, right_index=True)
```

- The exact columns to join on can also be specified. 

```python
spy['Name'] = 'SPY'
aapl['Name'] = 'AAPL'
merged_df = spy.merge(aapl, on=['Name', 'Adj Close', 'Volume'], how='outer')
merged_df
```

###  Join

- `pd.join` method can be used to join based on row labels rather than columns. 

```python
spy.join(aapl, lsuffix='_x', rsuffix='_y')
```

##  Handling NaNs

- NaNs (Not a Number) are different from Python's None type which also represent missing data in Python objects. 
- Most of the calculation by default will ignore rows with NaN values. 
- However, often you may have to specify in your operation, how you want the NaNs values to be dealt with. 
- In other cases, you have to remove them from the dataset.

```python
print(np.nan)
print(np.nan + 1)
print(np.nan * 0)
```

- `pd.isna` and `pd.isnull` methods check NaN condition.

```python
merged_df.isna()                    # boolean mask for NaN values
```

```python
merged_df.isnull().sum()             # count of no. of nan values by columns
```

```python
merged_df.dropna(axis=0)             # remove NaN values by rows
```

```python
merged_df.dropna(axis=1)             # remove NaN values by columns
```

- It is also possible to replace NaN values with some other values. This is called data imputation and useful when you face data scarcity problems.

```python
mu = merged_df['Adj Close'].mean()
merged_df['Adj Close'].fillna(mu)        # replace NA values with mean for Adj_Close column
```

##  Groupby

- Groupby operations involves splitting the dataset based on some grouping, applying same function to each of those groups separately and finally, combining the results. This can be performed using `pd.groupby` method.

```python
spy.head()
```

```python
concat_df = pd.concat( [spy, aapl] )
grouped = concat_df.groupby('Name')
grouped
```

```python
grouped.groups
```

```python
grouped.get_group('AAPL')
```

```python
grouped.mean()
```

- Groups can be formed based on multiple coulumns as well.

```python
concat_df 
```

```python
concat_df['Year'] = [i.year for i in concat_df.index]
concat_df['Month'] = [i.month for i in concat_df.index]
concat_df
```

```python
mean_df = concat_df.groupby( ['Name', 'Year', 'Month'] ).mean()
mean_df
```

##  Applying and Lambda Function

- `lambda` function is a short way of defining a function where the `def` and `return` keywords are not needed.

```python
twosum = lambda a,b:a+b
twosum(1,2)
```

- It is possible to use the `lambda` function with `pd.apply` method and apply same the function to each row of each column, given the operation is possible in the column.

```python
df=pd.DataFrame(np.arange(1,13).reshape(4,3)) 
df 
```

```python
df.apply( lambda i:i+10 )
```

```python
df.apply(lambda i : i>10)
```

- `pd.apply` method is most useful when you want to apply your own functions to a dataframe.

```python
def columndiff(col1, col2):
    return(col2 - col1)
```

```python
df['diff'] = df.apply( lambda x : columndiff(x[2], x[1]), axis=1) 
df.head()
```

```python
plt.scatter(x, y, marker='*', alpha=.25)

plt.title("AAPL vs MSFT", fontsize=16)

plt.ylabel('AAPL', fontsize=14)
plt.xlabel('MSFT', fontsize=14)

plt.xlim(100, 250)
plt.ylim(100, 400)

plt.show()
```

- Multiple items can be plotted in the same graph. 
- `plt.xticks`, `plt.yticks` allows you to adjust how the tick labels of the x and y axis are displayed. 
- `plt.legend` allows you to display and customize legends of graphs.

```python
name = 'AAPL'
mov_avg = stocks['AAPL'].rolling(window=10).mean()

stocks['AAPL'].plot(figsize=(10,5), label=name)
mov_avg.plot(label='moving avg', linestyle='dashed')

plt.title("Adjusted Close Price of %s" % name, fontsize=16)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)

plt.xticks(rotation=90)
plt.legend([name,'moving avg'], frameon=True, shadow=True)

plt.show()
```

## Multiple Plots

- Multiple figures can be plotted in the same figure.

```python
fig = plt.figure()
rect=[0.1,0.2,0.8,0.9]
ax1=fig.add_axes(rect)

rect=[0.2,0.3,0.3,0.5]
ax2=fig.add_axes(rect)
```

- Using `plt.subplots` method, multiple plots can also be plotted in the same figure. 

```python
fig, ax = plt.subplots(1,3)    # create plotting space of 1 row and 3 columns
plt.show()
```

- When plotting more than one plot at once, the size of the figure of the entire plotting space can be controlled using `mpl.rc` method. 

```python
mpl.rc('figure', figsize=(10,4))
fig, ax = plt.subplots(1,3)    # create plotting space of 1 row and 3 columns
plt.show()
```

- `plt.subplots` creates space for plotting each plot. It ouputs a tuple with two items - 
    - the figure space where the plots are placed
    - an array, which can be indexed to place the desired plot.
- `.set_title` allows you to place title for each plot.
- `plt.suptitle` gives title to the entire figure.

```python
mpl.rc('figure', figsize=(10,4))
fig, ax = plt.subplots(1,3)

ax[0].scatter(stocks.AAPL.values, stocks.MSFT.values,  alpha=.25)
ax[0].set_title('AAPL vs MSFT')

ax[1].scatter(stocks.AAPL.values, stocks.GOOG.values, c='y', alpha=.5)
ax[1].set_title('AAPL vs GOOG')

ax[2].scatter(stocks.MSFT.values, stocks.GOOG.values, c='#FF0000',  alpha=.75)
ax[2].set_title('MSFT vs GOOG')

plt.suptitle('Scatter plot of AAPL, MSFT and GOOG')
plt.show()
```

- `plt.style.use` can be used to change the background of a plot.
- `.set` method can be used to label axes of each plot. 

```python
plt.style.use('ggplot')
mpl.rc('figure', figsize=(15,4))
fig, ax = plt.subplots(1,3)

ax[0].scatter(stocks.AAPL.values, stocks.MSFT.values, c='blue')
ax[1].scatter(stocks.AAPL.values, stocks.GOOG.values, c='y')
ax[2].scatter(stocks.MSFT.values, stocks.GOOG.values, c='#FF0000')

ax[0].set(xlabel='AAPL',ylabel='MSFT')
ax[1].set(xlabel='AAPL',ylabel='GOOG')
ax[2].set(xlabel='MSFT',ylabel='GOOG')

plt.show()
```
